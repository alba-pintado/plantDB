## <font color='FIREBRICK'>Ejercicio 1 (10 puntos)</font>

<font size='3px'><p style="text-align: justify;line-height: 150%">Un laboratorio de botánica quiere organizar su colección de plantas en una base de datos (BBDD) a fin de proporcionar información adecuada a sus investigadores. No quiere utilizar ninguna BBDD existente y nos propone desarrollarla desde cero:</p></font>
    
<font size='3px'><p style="text-align: justify;line-height: 150%">La BBDD será un diccionario de diccionarios. Para el primer diccionario la clave será el nombre de la planta. El segundo diccionario (el que estaría dentro del primer diccionario) consta de 3 campos: tamaño en centímetros de la planta, hábitat (acuático o terrestre) y una lista con las provincias donde puede cultivarse.</p></font>

<font size='3px'><p style="text-align: justify;line-height: 150%">El programa nos permitirá hacer una serie de cosas, elegidas por un menú:</p></font>

<div style="font-size: 16px; line-height: 150%">
<ol>
    1. Introducir los datos de una nueva planta (el número de provincias donde se puede cultivar puede variar de una a otra) (Si ya existía en la BBDD se sobreescribirán los datos con los nuevos).<br/>
    2. Listar todas las plantas mostrando únicamente sus nombres (con cierto formato).<br/>
    3. Mostrar los datos de una planta determinada (con cierto formato).<br/>
    4. Buscar aquellas plantas cuyo tamaño se encuentre en un determinado rango que se pedirá al usuario (p.ej entre 120 y 180cm).<br/>
    5. Buscar aquellas plantas de un hábitat determinado (Acuático/terrestre).<br/>
    6. Buscar aquellas plantas que se puedan cultivar en una provincia.<br/>
    0. Salir del programa.        
</ol>
</div>

<font size='3px'><p style="text-align: justify;line-height: 150%">El programa debe estructurarse con funciones para cada una de las diferentes opciones. Las funciones deberán recibir como argumento la BBDD. Pedirán al usuario la información necesaria, y la procesarán para devolver el resultado pertinente. La función principal del programa (o "main") que contenga el menú, será interactiva para el usuario, permitiendo ejecutar varias funcionalidades hasta que el usuario quiera salir.</p></font>

<font size='3px'><p style="text-align: justify;line-height: 150%"><font size='3px'><p style="text-align: justify;line-height: 150%">Para no perder los datos que el usuario va introduciendo vamos a almacenar la BBDD en un fichero llamado 'BBDD_plantas.txt'. De esta forma, cada vez que se ejecute el programa deberán recuperarse los datos almacenados previamente y al salir del programa debe escribirse la información de la BBDD en fichero. Si al iniciar el programa, no existe el fichero de BBDD, deberá crearse y advertir al usuario de que comienza con una BBDD vacía.</p></font></p></font>

```python
#(1) Importación de módulos:
# En Python, un módulo es un archivo que contiene funciones, clases y herramientas ya preparadas para usarse.
# En lugar de programarlo todo desde cero, Python permite importar módulos para aprovechar código ya existente.

# Importa el módulo 'json' para leer y escribir archivos JSON.
# Permite trabajar con archivos y datos en formato JSON (JavaScript Object Notation), un formato de texto emplado para guardar datoss, enviar información entre programa y configurar archivos.

import json  

# Importa el módulo 'os' para verificar existencia de ficheros y manejar rutas
# Permite interactuar con el sistema operativo, en nuestro caso, LINUX.
import os    


# (2) Función para cargar la BBDD:

# Define la función cargar_bbdd que recibe un nombre de fichero por defecto
def cargar_bbdd(nombre_fichero='BBDD_plantas.txt'):
    
    # Comprueba si el fichero existe
    if os.path.exists(nombre_fichero):  
        # Abre el fichero en modo lectura con codificación UTF-8
        with open(nombre_fichero, 'r', encoding='utf-8') as f:  
            try:
                
                # Intenta cargar los datos JSON del fichero
                bbdd = json.load(f)  
                # Si el fichero existe pero está vacío, muestra un mensaje
                if not bbdd:  
                    print("El fichero existe pero está vacío. Se empieza con una BBDD vacía.")
           
          # Captura error si el JSON está dañado
            except json.JSONDecodeError:  
                print("El fichero estaba dañado. Se inicia una BBDD vacía.")
                bbdd = {}
    else:
        # Si el fichero no existe, inicia una BBDD vacía y muestra un mensaje
        print("No existe la BBDD. Se iniciará una base de datos vacía.")
        bbdd = {}
    
    # Devuelve la BBDD cargada
    return bbdd  

# (3) Función para guardar la BBDD:

# Define la función guardar_bbdd que recibe la BBDD y el nombre del fichero
def guardar_bbdd(bbdd, nombre_fichero='BBDD_plantas.txt'):
    # Abre el fichero en modo escritura con codificación UTF-8
    with open(nombre_fichero, 'w', encoding='utf-8') as f:  
        # Guarda la BBDD en formato JSON con indentación y sin caracteres ASCII especiales
        json.dump(bbdd, f, indent=4, ensure_ascii=False)  

# (4) Función para introducir o actualizar una planta:

# Define la función introducir_planta que recibe la BBDD
def introducir_planta(bbdd):
    # Solicita al usuario el nombre de la planta y elimina espacios
    nombre = input("Nombre de la planta: ").strip()  
    
    # Bucle para validar que el tamaño sea un número positivo
    while True:  
        try:
            # Solicita el tamaño de la planta
            tamaño = float(input("Tamaño en cm (número positivo): "))  
            # Comprueba que sea mayor a 0
            if tamaño <= 0:  
                print("El tamaño debe ser positivo.")
                continue
            # Sale del bucle si es válido
            break  
        except ValueError:  
            # Mensaje si el usuario no introduce un número válido
            print("Debes introducir un número válido.")
    
    # Bucle para validar el hábitat
    while True:  
        # Solicita el hábitat y lo convierte a minúsculas
        habitat = input("Hábitat de la planta (acuático/terrestre): ").strip().lower()
        if habitat in ["acuatico", "acuático"]:
            habitat = "acuático"
            break
        elif habitat == "terrestre":
            break
        else:
            # Mensaje si no es válido
            print("Debes escribir 'acuático' o 'terrestre'.")
    
    # Lista oficial de provincias españolas
    PROVINCIAS_ES = [
        "Álava","Albacete","Alicante","Almería","Asturias","Ávila","Badajoz","Barcelona",
        "Burgos","Cáceres","Cádiz","Cantabria","Castellón","Ciudad Real","Córdoba","La Coruña",
        "Cuenca","Gerona","Granada","Guadalajara","Guipúzcoa","Huelva","Huesca","Islas Baleares",
        "Jaén","León","Lérida","Lugo","Madrid","Málaga","Murcia","Navarra","Orense","Palencia",
        "Las Palmas","Pontevedra","La Rioja","Salamanca","Santa Cruz de Tenerife","Segovia","Sevilla",
        "Soria","Tarragona","Teruel","Toledo","Valencia","Valladolid","Vizcaya","Zamora","Zaragoza"
    ]
    
    # Bucle para validar las provincias introducidas por el usuario
    while True:  
        # Solicita al usuario las provincias donde se puede cultivar
        prov_input = input("Introduce la/las provincia/s donde se cultiva principalmente la especie indicada (separadas por comas, respetando mayúsculas y tildes): ").strip()
        # Crea una lista de provincias eliminando espacios vacíos
        provincias = [p.strip() for p in prov_input.split(",") if p.strip()]  
        # Comprueba si hay provincias inválidas
        invalidas = [p for p in provincias if p not in PROVINCIAS_ES]  
        if invalidas:
            # Mensaje indicando provincias inválidas
            print()
            print("Estas provincias no son válidas o están mal escritas:", ", ".join(invalidas))
            print("Escribe correctamente las provincias, respetando mayúsculas iniciales y tildes.")
        else:
            # Sale del bucle si todas son válidas
            break
    
    # Si la especie de planta ya existe, avisa quelos cambios realizados sobre la misma se actualizarán y modificarán
    if nombre in bbdd:  
        print(f"La planta '{nombre}' ya existía y se actualizará.")
    
    # Añade o actualiza la planta en la BBDD
    bbdd[nombre] = {"tamaño": tamaño, "hábitat": habitat, "provincias": provincias}  
    print(f"\nLa planta '{nombre}' ha sido añadida a la BBDD correctamente.\n")
    print("Recuerde, para guardar los cambios realizados, debe seleccionar la opción 0.\n")

# (5) Función para listar todas las plantas:

# Define la función listar_plantas que recibe la BBDD
def listar_plantas(bbdd):
    
    # Si la BBDD está vacía, muestra un mensaje
    if not bbdd:  
        print("No hay plantas en la BBDD.")  
        return
    # Muestra un título
    print("Lista de plantas añadidas a la BBDD (orden alfabético):")
    
    # Recorre la BBDD ordenada alfabéticamente y enumera cada especie
    for i, nombre in enumerate(sorted(bbdd.keys()), start=1):  
        print(f"{i}. {nombre}")
    # Si hay al menos una planta, muestra la última añadida
    if bbdd:
        ultima = list(bbdd.keys())[-1]  # Obtiene la última clave añadida
        print(f"\nÚltima especie añadida a la BBDD: {ultima}")
    print()
    

# (6) Función para mostrar los datos de una planta:

# Define la función mostrar_planta que recibe la BBDD
def mostrar_planta(bbdd):
    # Solicita el nombre de la planta
    nombre = input("Introduce el nombre de la planta a mostrar: ").strip()
    # Si la planta existe en la BBDD
    if nombre in bbdd:  
        # Obtiene los datos de la planta
        datos = bbdd[nombre]  
        # Muestra los datos
        print(f"Nombre: {nombre}")
        print(f"Tamaño: {datos['tamaño']} cm")
        print(f"Hábitat: {datos['hábitat']}")
        print(f"Provincias: {', '.join(datos['provincias'])}\n")
    else:
        # Mensaje si no existe
        print(f"No existe la planta '{nombre}' en la BBDD.")  

# (7) Función de búsqueda por tamaño mostrando nombre y tamaño:

# Define la función buscar_por_tamaño que recibe la BBDD
def buscar_por_tamaño(bbdd):
    
    # Solicita el rango de tamaños sobre el que se quiere buscar la o las especies a estudiar
    try:
        min_t = float(input("Tamaño mínimo en cm: "))  
        max_t = float(input("Tamaño máximo en cm: "))  
    except ValueError:
        print("Debes introducir números válidos.")
        return
    
    # Filtra las plantas que estén dentro del rango
    resultados = [(n, d['tamaño']) for n, d in bbdd.items() if min_t <= d['tamaño'] <= max_t]  
    if resultados:
        
        # Ordena por tamaño y alfabéticamente en caso de empate
        resultados.sort(key=lambda x: (x[1], x[0]))
        print()
        print(f"Especies en el rango solicitado ({min_t} - {max_t} cm) (nombre - tamaño en cm):")
        print()
      # Enumera resultados
        for i, (nombre, tamaño) in enumerate(resultados, start=1):  
            print(f"{i}. {nombre} - {tamaño} cm")
    
    else:
        print("No se encontraron plantas en ese rango.")
    
    # Nota aclatoria sobre el orden de los resultados que se muestran al usuario
    print()
    print("Nota: Los resultados se muestran de menor a mayor tamaño (cm). Si hay tamaños iguales, la especies se muestran en orden alfabético.")

    # (8) Función de búsqueda por hábitat mostrando solo nombres ordenados y enumerados:

# Define la función buscar_por_habitat que recibe la BBDD
def buscar_por_habitat(bbdd):
    # Solicita el hábitat a buscar
    habitat = input("Hábitat a buscar (acuático/terrestre): ").strip().lower()
    # Normaliza el nombre
    if habitat in ["acuatico", "acuático"]:
        habitat = "acuático"
    elif habitat != "terrestre":
        print("Hábitat no válido.")
        return
    # Filtra las especies que tienen ese hábitat
    resultados = [n for n, d in bbdd.items() if d['hábitat'] == habitat]  
    if resultados:
        # Ordena alfabéticamente
        resultados.sort()
        print()
        print(f"Especies de hábitat '{habitat}':")
        # Enumera resultados
        for i, nombre in enumerate(resultados, start=1):  
            print(f"{i}. {nombre}")
    else:
        print(f"No se encontraron plantas de hábitat '{habitat}'.")
    
    print()
    print(f"Nota: Los resultados se muestran en orden alfabético.")
    
# (9) Función de búsqueda por provincia mostrando resultados enumerados y ordenados alfabéticamente:

# Define la función buscar_por_provincia que recibe la BBDD
def buscar_por_provincia(bbdd):
    # Lista oficial de provincias
    PROVINCIAS_ES = [
        "Álava","Albacete","Alicante","Almería","Asturias","Ávila","Badajoz","Barcelona",
        "Burgos","Cáceres","Cádiz","Cantabria","Castellón","Ciudad Real","Córdoba","La Coruña",
        "Cuenca","Gerona","Granada","Guadalajara","Guipúzcoa","Huelva","Huesca","Islas Baleares",
        "Jaén","León","Lérida","Lugo","Madrid","Málaga","Murcia","Navarra","Orense","Palencia",
        "Las Palmas","Pontevedra","La Rioja","Salamanca","Santa Cruz de Tenerife","Segovia","Sevilla",
        "Soria","Tarragona","Teruel","Toledo","Valencia","Valladolid","Vizcaya","Zamora","Zaragoza"
    ]
    while True:  # Valida provincia
        # Solicita provincia al usuario
        provincia = input("Provincia a buscar: ").strip()
        if provincia not in PROVINCIAS_ES:
            print()
            print(f"La provincia '{provincia}' no es válida. Escribe correctamente respetando mayúscula inicial y tildes.")
        else:
            break
    # Filtra plantas que se pueden cultivar en esa provincia
    resultados = [n for n, d in bbdd.items() if provincia in d['provincias']]  
    if resultados:
        # Ordena alfabéticamente
        resultados.sort()  
        print(f"Especies que se cultivan principalmente en '{provincia}':")
        print()
        # Enumera resultados
        for i, nombre in enumerate(resultados, start=1):  
            print(f"{i}. {nombre}")
    else:
        print(f"No se encontraron plantas para la provincia '{provincia}'.")
    print()
    print(f"Nota: Los resultados se muestran en orden alfabético.")
    
# (10) Función principal con menú:

# Define la función main para ejecutar el programa
def main():
    # Carga la BBDD al iniciar
    bbdd = cargar_bbdd()  
    while True:  # Bucle principal del menú
        # Muestra opciones del menú
        print("\nMenú principal")
        print()
        print("1. Introducir nueva especie de planta (Nombre científico (Nombre común)):")
        print("2. Listar todas las especies de plantas")
        print("3. Mostrar datos de una especie de planta")
        print("4. Buscar especies de plantas por tamaño")
        print("5. Buscar especies de plantas por hábitat")
        print("6. Buscar especies de plantas por provincia")
        print("0. Salir de la BBDD")
        print()
        # Nota sobre guardar cambios
        print("Importante: cualquier planta nueva añadida se guardará en la BBDD solo al salir seleccionando la opción 0. De no ser asi, los datos introducidos no se guardaran.\n")
        # Solicita opción al usuario
        opcion = input("Elija una opción (0-6): ").strip()
        print()
        # Llama a la función correspondiente según la opción
        if opcion == "1":
            introducir_planta(bbdd)
        elif opcion == "2":
            listar_plantas(bbdd)
        elif opcion == "3":
            mostrar_planta(bbdd)
        elif opcion == "4":
            buscar_por_tamaño(bbdd)
        elif opcion == "5":
            buscar_por_habitat(bbdd)
        elif opcion == "6":
            buscar_por_provincia(bbdd)
        elif opcion == "0":
            # Guarda la BBDD antes de salir
            guardar_bbdd(bbdd)  
            print("La BBDD ha sido actualizada.")
            print("Gracias por su aportación. Nos vemos muy pronto. ¡Hasta luego!")
            break
        else:
            #Mensaje para el usuario de opción incorrecta
            print("Opción no válida.")  
            
# (11) Ejecutar el programa
# Llama a la función principal para iniciar el programa
main()  
```
